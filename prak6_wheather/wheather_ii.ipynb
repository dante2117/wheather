```python
import pandas as pd
import numpy as np #подключение библиотек
```

```python
df = pd.read_csv('weatherAUS_clear.csv') #выгрузка данных в датасет
df
```

```python
df = df.drop(columns='Unnamed: 0')
df = df.drop(columns='Date')
df #удаление столбцов
```

```python
from sklearn.preprocessing import OrdinalEncoder #подключение библиотеки

Oenc = OrdinalEncoder()
Oenc.fit(df[['Location','WindGustDir']]) #установка значения
df
```

```python
Oenc.categories_ #ожидаемые категории
```

```python
Oenc = OrdinalEncoder()
df[['Location','WindGustDir']] = Oenc.fit_transform(df[['Location','WindGustDir']]) #установка значений и их преобразование
df
```

```python
from sklearn.preprocessing import LabelEncoder #подключение библиотеки
Lenc = LabelEncoder()
Lenc.fit(df['RainTomorrow']) #установка кодировщика меток
df['RainTomorrow'] = Lenc.fit_transform(df['RainTomorrow']) #возврат закодированных меток
df
```

```python
Lenc.classes_ #метки для всех классов
```

```python
from sklearn.preprocessing import OneHotEncoder #подключение библиотеки
Onhot = OneHotEncoder()
Onhot.fit_transform(df[['RainToday']]).toarray() #совоставление данных и преобразование в массив
```

```python
df[['RainToday', 'RainTodayYes']] = Onhot.fit_transform(df[['RainToday']]).toarray() #совоставление данных и преобразование в массив по столбцам
df
```

```python
df = df.drop(columns='RainToday') #удаление столбца
df
```

```python
from sklearn.preprocessing import Normalizer #импорт библиотеки

Norm = Normalizer()

df[['Pressure9am', 'Pressure3pm']] = Norm.fit_transform(df[['Pressure9am', 'Pressure3pm']]) #преобразование данных числовых столбцов
df
```

```python
from sklearn.model_selection import train_test_split #импорт библиотеки для разбития на подмножества

Y = df['RainTomorrow'] #присвоение х и у
X = df.drop(columns=['RainTomorrow'])

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42) 
#формирование разделённых данных с размером текста и перетасовкой данных int
```
